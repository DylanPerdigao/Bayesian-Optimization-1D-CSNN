```python
import optuna
import pandas as pd

pd.set_option('display.max_columns', 100)
```

```python
configs = [
    ('Net1_CSNN-Base-E5-B128-S20'),
    ('Net1_CSNN-Base-E5-B512-S20'),  
    ('Net1_CSNN-Base-E5-B1024-S20'),
    ('Net1_CSNN-Base-E10-B128-S20'), 
    ('Net1_CSNN-Base-E10-B512-S20'),
    ('Net1_CSNN-Base-E10-B1024-S20'),
    ('Net1_CSNN-Base-E5-B1024-S50'), 
    ('Net1_CSNN-Base-E5-B1024-S10'),

    ('Net2_CSNN-Base-E5-B128-S20'),
    ('Net2_CSNN-Base-E5-B512-S20'),
    ('Net2_CSNN-Base-E5-B1024-S20'),
    ('Net2_CSNN-Base-E10-B128-S20'),
    ('Net2_CSNN-Base-E10-B512-S20'),
    ('Net2_CSNN-Base-E10-B1024-S20'),
    ('Net2_CSNN-Base-E5-B1024-S50'), 
    ('Net2_CSNN-Base-E5-B1024-S10'),

    ('Net3_CSNN-Base-E5-B128-S20'),
    ('Net3_CSNN-Base-E5-B512-S20'),
    ('Net3_CSNN-Base-E5-B1024-S20'),
    ('Net3_CSNN-Base-E10-B128-S20'),
    ('Net3_CSNN-Base-E10-B512-S20'),
    ('Net3_CSNN-Base-E10-B1024-S20'),
    ('Net3_CSNN-Base-E5-B1024-S50'), 
    ('Net3_CSNN-Base-E5-B1024-S10'),
]
```

```python
best_trials = pd.DataFrame()
for study_name in configs:
    study = optuna.load_study(study_name=study_name, storage='sqlite:///../epia2024.db')
    df = study.trials_dataframe()
    df = df[df['values_0'] <= 0.05]
    print(df.shape, study_name)
    try:
        df = df.sort_values(by='user_attrs_@5FPR recall', ascending=False)
    except:
        print(f'{study_name} not found')
    df = df.head(1)
    df['net'] = f"Net{study_name.split('Net')[1].split('_')[0]}"
    df['epochs'] = int(study_name.split('E')[1].split('-')[0])
    df['batch_size'] = int(study_name.split('B')[2].split('-')[0])
    df['steps'] = int(study_name.split('S')[2].split('-')[0])
    best_trials = pd.concat([best_trials, df])
best_trials = best_trials.sort_values(by='user_attrs_@5FPR recall', ascending=False)
best_trials
```

```python
best_trials['user_attrs_@5FPR f1_score'] = 2 * (best_trials['user_attrs_@5FPR recall'] * best_trials['user_attrs_@5FPR precision']) / (best_trials['user_attrs_@5FPR recall'] + best_trials['user_attrs_@5FPR precision'])
```

```python
best_trials_selected = best_trials[['net', 'epochs', 'batch_size', 'steps', 'user_attrs_@5FPR recall', 'user_attrs_@5FPR fpr', 'user_attrs_@5FPR fpr_ratio']]
best_trials_selected_sorted = best_trials_selected.sort_values(by=['net', 'epochs', 'batch_size', 'steps'])
best_trials_selected_sorted
```

```python
best_trials_selected_sorted['user_attrs_@5FPR recall'] = (best_trials_selected_sorted['user_attrs_@5FPR recall'] * 100).round(2)
best_trials_selected_sorted['user_attrs_@5FPR fpr'] = (best_trials_selected_sorted['user_attrs_@5FPR fpr'] * 100).round(2)
best_trials_selected_sorted['user_attrs_@5FPR fpr_ratio'] = (best_trials_selected_sorted['user_attrs_@5FPR fpr_ratio'] * 100).round(2)
best_trials_selected_sorted
```
